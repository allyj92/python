```python
import seaborn as sns
```

```python
iris = sns.load_dataset("iris")
sns.distplot(iris["sepal_width"],
            hist=True,
            kde=True,
            rug =True)
```

```python
tips = sns.load_dataset("tips")
sns.relplot(x = "total_bill",y="tip",hue="smoker",data=tips,kind ="line")
```

```python
sns.relplot(x = "total_bill",y="tip",hue="smoker",data=tips)
```

```python
sns.catplot(x="day",y="total_bill",data=tips)
sns.catplot(x="day",y="total_bill",kind="bar",data=tips)
sns.catplot(x="day",y="total_bill",kind="violin",data=tips)
```

```python
sns.boxplot(x="day",y="total_bill",data=tips)
```

```python
sns.jointplot(x="day",y="total_bill",data=tips)
```

```python
sns.jointplot(x="total_bill",y="tip",data=tips,kind="hex")
```

```python
iris = sns.load_dataset("iris")
sns.pairplot(iris)
sns.pairplot(iris,vars=["sepal_length","sepal_width"])
```

```python
flights=sns.load_dataset('flights')
flights = flights.pivot("month","year","passengers")
sns.heatmap(flights,annot=True,fmt="d")
```
